`Number Recognition `: Handwritten digit recognition system not only detects scanned images of handwritten digits.Handwritten digit recognition using MNIST dataset is a major project made
with the help of Neural Network. It basically detects the scanned images of handwritten digits.

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np

import keras
```

```python
# Load the MNIST dataset
(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.mnist.load_data()

```

```python
# Preprocess the data
train_images = train_images.reshape(60000, 28, 28, 1)
train_images = train_images.astype('float32') / 255
test_images = test_images.reshape(10000, 28, 28, 1)
test_images = test_images.astype('float32') / 255

```

```python
# One-hot encode the labels
train_labels_one_hot = keras.utils.to_categorical(train_labels)
test_labels_one_hot = keras.utils.to_categorical(test_labels)

```

```python

# Create the model
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(10, activation='softmax')
])
```

```python

# Train the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(train_images, train_labels_one_hot, epochs=5)

```

```python
# Evaluate the model
test_loss, test_accuracy = model.evaluate(test_images, test_labels_one_hot)
print('Test loss:', test_loss)
print('Test accuracy:', test_accuracy)
```

```python
model.save('mnist_model.h5')
```
